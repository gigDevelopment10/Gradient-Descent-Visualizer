**Gradient Descent** is an optimization algorithm used to minimize a function by iteratively adjusting its parameters in the direction of the steepest descent, as determined by the function's gradient (or derivative). It's widely used in machine learning and deep learning for minimizing the loss function, which measures how well a model is performing.


```python
import torch
import matplotlib.pyplot as plt
import numpy as np

# Create the loss function (quadratic) L(x, y) = x^2 + y^2
def loss_function(x, y):
    return x**2 + y**2

# Define the gradient descent process
def gradient_descent(learning_rate, num_iterations):
    # Initialize parameters randomly
    x = torch.tensor([2.0], requires_grad=True)  # Initial x
    y = torch.tensor([2.0], requires_grad=True)  # Initial y

    # Store the path of updates
    path_x = [x.item()]
    path_y = [y.item()]
    losses = [loss_function(x, y).item()]

    for i in range(num_iterations):
        # Compute the loss
        loss = loss_function(x, y)

        # Compute gradients
        loss.backward()

        # Update parameters using gradient descent
        with torch.no_grad():
            x -= learning_rate * x.grad
            y -= learning_rate * y.grad

        # Zero the gradients after each step
        x.grad.zero_()
        y.grad.zero_()

        # Store the updated parameters and loss
        path_x.append(x.item())
        path_y.append(y.item())
        losses.append(loss.item())

    return path_x, path_y, losses

# Gradient descent parameters
learning_rate = 0.1
num_iterations = 50

# Run gradient descent
path_x, path_y, losses = gradient_descent(learning_rate, num_iterations)

# Visualization of the Loss Landscape and the gradient descent path
x_vals = np.linspace(-2.5, 2.5, 400)
y_vals = np.linspace(-2.5, 2.5, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = X**2 + Y**2

# Plot the contour map
plt.figure(figsize=(8,6))
plt.contour(X, Y, Z, levels=50, cmap='jet')

# Plot the gradient descent path
plt.plot(path_x, path_y, marker='o', color='red')
plt.scatter(path_x, path_y, c='red', s=10)
plt.title('Gradient Descent on Loss Landscape')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import torch
import matplotlib.pyplot as plt
import numpy as np

# Define the gradient descent process to accept a custom loss function
def gradient_descent(loss_function, learning_rate, num_iterations):
    # Initialize parameters randomly
    x = torch.tensor([2.0], requires_grad=True)  # Initial x
    y = torch.tensor([2.0], requires_grad=True)  # Initial y

    # Store the path of updates
    path_x = [x.item()]
    path_y = [y.item()]
    losses = [loss_function(x, y).item()]

    for i in range(num_iterations):
        # Compute the loss
        loss = loss_function(x, y)

        # Compute gradients
        loss.backward()

        # Update parameters using gradient descent
        with torch.no_grad():
            x -= learning_rate * x.grad
            y -= learning_rate * y.grad

        # Zero the gradients after each step
        x.grad.zero_()
        y.grad.zero_()

        # Store the updated parameters and loss
        path_x.append(x.item())
        path_y.append(y.item())
        losses.append(loss.item())

    return path_x, path_y, losses
```

```python
# Example of using a different loss function: Rosenbrock function
def rosenbrock_loss(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2

# Gradient descent parameters
learning_rate = 0.001
num_iterations = 2000

# Run gradient descent with the Rosenbrock function
path_x, path_y, losses = gradient_descent(rosenbrock_loss, learning_rate, num_iterations)

# Visualization of the Loss Landscape and the gradient descent path
x_vals = np.linspace(-2.5, 2.5, 400)
y_vals = np.linspace(-2.5, 2.5, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = (1 - X)**2 + 100 * (Y - X**2)**2  # Rosenbrock landscape

# Plot the contour map
plt.figure(figsize=(8,6))
plt.contour(X, Y, Z, levels=50, cmap='jet')

# Plot the gradient descent path
plt.plot(path_x, path_y, marker='o', color='red')
plt.scatter(path_x, path_y, c='red', s=10)
plt.title('Gradient Descent on Rosenbrock Loss Landscape')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Example of using a different loss function: Himmelblau's function
def himmelblau_loss(x, y):
    return (x**2 + y - 11)**2 + (x + y**2 - 7)**2

# Gradient descent parameters
learning_rate = 0.001
num_iterations = 2000

# Run gradient descent with the Himmelblau function
path_x, path_y, losses = gradient_descent(himmelblau_loss, learning_rate, num_iterations)

# Visualization of the Loss Landscape and the gradient descent path
x_vals = np.linspace(-2.5, 2.5, 400)
y_vals = np.linspace(-2.5, 2.5, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = (1 - X)**2 + 100 * (Y - X**2)**2  # Himmelblau landscape

# Plot the contour map
plt.figure(figsize=(8,6))
plt.contour(X, Y, Z, levels=50, cmap='jet')

# Plot the gradient descent path
plt.plot(path_x, path_y, marker='o', color='red')
plt.scatter(path_x, path_y, c='red', s=10)
plt.title('Gradient Descent on Himmelblau Loss Landscape')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```


**How These Functions Differ:**

*  Rosenbrock: Creates a long, narrow valley, making it hard for gradient descent to converge unless carefully tuned.

*  Himmelblau: Has multiple local minima, and gradient descent may fall into different minima depending on the initial values.

*  Beale: Has a more complicated shape with a steep region near the minimum.
